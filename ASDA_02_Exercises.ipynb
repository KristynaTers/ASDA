# Exercise 2 — Manual t-test

An OLS regression gives β̂ = 0.85, SE(β̂) = 0.12, n = 80.

Compute the t-statistic and test H₀: β = 0 at the 5% significance level.

````python
beta_hat = 0.85
se_beta  = 0.12
n        = 80
beta_0   = 0  # null hypothesis value

# YOUR CODE HERE
# t_stat = ...
# t_crit = ...  (use stats.t.ppf)
# p_val  = ...
# print decision
````

<details><summary>💡 Solution</summary>

```python
t_stat = (beta_hat - beta_0) / se_beta  # = 7.083
t_crit = stats.t.ppf(0.975, df=n-2)    # ≈ 1.991
p_val  = 2*(1-stats.t.cdf(abs(t_stat), df=n-2))
print(f't-stat = {t_stat:.3f}, t* = {t_crit:.3f}, p = {p_val:.6f}')
print('Decision: REJECT H₀ — β is statistically significant')
```

# Exercise 7 — Simulate the t-distribution

Build an intuition for t-tests by running 1,000 simulations.

````python
np.random.seed(42)
N_SIM = 1000; N_OBS = 100
TRUE_BETA = 0  # H0 is TRUE

t_stats = []
for _ in range(N_SIM):
    x = np.random.normal(0, 1, N_OBS)
    y = TRUE_BETA * x + np.random.normal(0, 1, N_OBS)
    X_sim = sm.add_constant(x)
    m = sm.OLS(y, X_sim).fit()
    t_stats.append(m.tvalues[1])

t_stats = np.array(t_stats)
rejection_rate = np.mean(np.abs(t_stats) > stats.t.ppf(0.975, df=N_OBS-2))

print(f'True beta: {TRUE_BETA}')
print(f'False rejection rate: {rejection_rate:.3f} (should be ~0.05)')

# YOUR CODE HERE: plot histogram of t_stats vs theoretical t-distribution
````
